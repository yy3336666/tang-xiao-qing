```python
#b
def OutlierDetection(df,ks_res):
    u = df['price_per_sqft'].mean()
    std = df['price_per_sqft'].std()
    if ks_res==1:
        error = df[np.abs(df['price_per_sqft'] - u) > 2* std]
        data_c = df[np.abs(df['price_per_sqft'] - u) <= 2 * std]
        return error
error=OutlierDetection(data,1)
new_data=data.drop(index=error.index)
new_data.index=range(len(new_data))
new_data# Deleted by the standard deviation method, by retaining the data within the range of two standard deviations before and after
```

```python
#c
n, bins, patches = plt.hist(new_data['price_per_sqft'],density=True, facecolor='blue', alpha=0.5)
y = norm.pdf(bins, new_data['price_per_sqft'].mean(), new_data['price_per_sqft'].std())
plt.plot(bins,y)
plt.show()
```

```python
#d
def z_score(df):
    p=df['price_per_sqft']
    df_zscores=(p-p.mean())/p.std()
    error=df[np.abs(df_zscores)>2.2]
    return error
error=z_score(data)
new_data1=data.drop(index=error.index)
new_data1.index=range(len(new_data1))
new_data1#Deleted by the z-score method, and retained data with a value within 2.2 by normalizing the data
```

```python
movies=pd.read_csv('movies.csv')
movies=movies.dropna()
plt.scatter(movies['budget'],movies['gross'])
plt.xlabel('budget')
plt.ylabel('gross')
plt.title('budget--gross')
```

```python
#b
model=LinearRegression()
model=model.fit(np.array(movies['budget']).reshape(-1,1),np.array(movies['gross']).reshape(-1,1))
plt.scatter(movies['budget'],movies['gross'])
plt.plot(np.array(movies['budget']).reshape(-1,1),model.predict(np.array(movies['budget']).reshape(-1,1)))
```

```python
#c
sns.jointplot(movies.loc[:,['budget','gross']])
# make sense
```

```python
#d
pearsonr(movies['budget'], movies['gross'])[0]
#The correlation is 0.7402465439219629
```

```python
#e
def kendalltau_movies(df):
    corr1=np.zeros((df.shape[1],df.shape[1]))
    for i in range(corr1.shape[0]):
        for j in range(corr1.shape[1]):
            corr1[i,j]=kendalltau(df.iloc[:,i],df.iloc[:,j])[0]
    return corr1
corr1=kendalltau_movies(movies)
print(corr1[12])
# By showing the correlation matrix, it is found that 'vote' is highly correlated with 'gross',the correlation is 0.553625276 ,Higher than 'budge' correlation with 'gross'

```

```python
#f
corr1=pd.DataFrame(corr1,columns=movies.columns,index=movies.columns)
sns.heatmap(corr1)
# 'gross' is most correlated with 'vote' and 'budget', and negatively correlated with 'rating' and 'genre'
```

```python
data1= pd.read_csv('masses.txt')
data1= pd.read_csv('masses.txt', na_values=['?'], names = ['B', 'age','shape', 
'margin', 'density', 'severity'])
```

```python
#a
data1.fillna(method='ffill',inplace=True)
data1_feature=data1.iloc[:,:-1]
data1_lable=data1.iloc[:,-1]
sta=StandardScaler()
data1_feature=sta.fit_transform(data1_feature)
```

```python
#b
tree=DecisionTreeClassifier()
accuracy=cross_val_score(tree, data1_feature, data1_lable, cv=10,scoring='accuracy')
accuracy
```

```python
#c
acc=[]
for i in range(1,10):
    knn=KNeighborsClassifier(n_neighbors=i)
    accuracy_knn=cross_val_score(knn, data1_feature, data1_lable, cv=10,scoring='accuracy')
    acc.append(max(accuracy_knn))
K=np.argmax(acc)+1
print(K)
#The optimal value of K is 4
```

```python
#d
def create_baseline():
    model = Sequential()
    model.add(Dense(60, input_dim=5,activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
Classifier=KerasClassifier(build_fn=create_baseline,nb_epoch=100,batch_size=5)
accuracy_keras=cross_val_score(Classifier, data1_feature, data1_lable, cv=10)
max(accuracy_keras)
```

```python
from keras import optimizers
def create_baseline1():
    model = Sequential()
    model.add(Dense(60, input_dim=5,activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    sgd = optimizers.gradient_descent_v2.SGD(lr=0.1, decay=1e-6, momentum=0.9, nesterov=True)
    model.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])
    return model
Classifier=KerasClassifier(build_fn=create_baseline,nb_epoch=100,batch_size=5)
accuracy_keras=cross_val_score(Classifier, data1_feature, data1_lable, cv=10)
max(accuracy_keras)
```

```python
def create_baseline1():
    model = Sequential()
    model.add(Dense(60, input_dim=5,activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
    return model
Classifier=KerasClassifier(build_fn=create_baseline,nb_epoch=100,batch_size=5)
accuracy_keras=cross_val_score(Classifier, data1_feature, data1_lable, cv=10)
max(accuracy_keras)
```
