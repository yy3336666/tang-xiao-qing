```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df1=pd.read_csv("E-CommerceSales\Sale Report.csv").iloc[:,2:]
df2=pd.read_csv('E-CommerceSales\International sale Report.csv').iloc[:,1:]
df3=pd.read_csv('E-CommerceSales\Amazon Sale Report.csv').iloc[:,1:]
```

```python
df3
```

```python
df1.dropna(inplace=True)
df2.dropna(inplace=True)#对三种表格进行删除缺失值
df3.dropna(inplace=True)
```

```python
df1.drop_duplicates(inplace=True)
df2.drop_duplicates(inplace=True)
df3.drop_duplicates(inplace=True)#对三种表格进行去重
```

```python
a=df1.groupby('Category')['Stock'].sum()
plt.xticks(rotation=90)
plt.bar(a.index,a)#显示销售种类和库存的关系，‘kurta'种类库存最多，接近1200000
```

```python
b=df2['Style'].value_counts()[:20]
plt.xticks(rotation=90)
plt.bar(b.index,b)#显示所有天数商品风格的数量，shipping 风格最多，最受欢迎。
```

```python
c=df3.groupby('Category')['Amount'].sum()
plt.xticks(rotation=90)
plt.bar(c.index,c)#显示销售种类和数量的关系，‘set'种类库存最多，接近6*1e6
```

```python

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cate_col1=df1.select_dtypes('object').columns
cate_col2=df2.select_dtypes('object').columns
cate_col3=df3.select_dtypes('object').columns


for  i in cate_col1:
    df1.loc[:,i]=le.fit_transform(df1.loc[:,i])
for  i in cate_col2:
    df2.loc[:,i]=le.fit_transform(df2.loc[:,i])
for  i in cate_col3:
    df3.loc[:,i]=le.fit_transform(df3.loc[:,i])#将所有字符串标签转化为数值标签
```

```python
df1_corr=df1.corr()
sns.heatmap(df1_corr)#展示第一个表格相关性的热图
```

```python
df2_corr=df2.corr()
sns.heatmap(df2_corr,annot=True)#展示第二个表格相关性的热图
```

```python
col=df3.nunique()[df3.nunique()!=1].index
df3=df3.loc[:,col]
df3_corr=df3.corr()
sns.heatmap(df3_corr)#展示第三个表格相关性的热图
```
